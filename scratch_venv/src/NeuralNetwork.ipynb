```python
import numpy as np
import pandas as pd
import os
import cv2 as cv
import matplotlib.pyplot as plt

BASE_PATH = "datasets/NeuralNetwork/CompleteImages/All data (Compressed)/"
labels = os.listdir(BASE_PATH)

sizes = []
for label in labels:
    print(label)
    label_path = "/".join([BASE_PATH, label])
    for path in os.listdir(label_path):
        sizes.append(cv.imread(label_path + "/" + path).shape)
        
plt.hist(sizes)
```
